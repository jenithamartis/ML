```python
#Importing Required Libraries
import pandas as pd
import sklearn as sklearn
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from six import StringIO
from sklearn.tree import export_graphviz
import graphviz
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import confusion_matrix,classification_report
```

```python
#Extracting the data from .csv file
data = pd.read_csv('CE802_P2_Data.csv')
data
```

```python
#Replacing the missing values using mean of the existing value
data.fillna(data.mean(),inplace=True)
data
```

```python
#Obtaining Class values
y= data.iloc[:,-1]
type(y)
```

```python
#Changing the type of y to DataFrame
y =pd.DataFrame(y)
#y.reset_index(drop=True)
type(y)
```

```python
#Obtaining feature values from F1 to F15
x=data.iloc[:,:-1]
type(x)
```

```python
print(x.shape)
print(y.shape)
```

```python
#Splitting the data into test and train
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.22, random_state=42)
```

```python
#Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn import tree
```

```python
#Creating a instance of Decision tree classifier
clf = tree.DecisionTreeClassifier()
clf
```

```python
#Training the model using train data
clf = clf.fit(X_train, y_train)
```

```python
#Predicting the output for X_test
dec_y_pred = clf.predict(X_test)
dec_y_pred.shape
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_train.values.flatten()
y_train.shape
```

```python
#Finding the accuracy score
print('Accuracy Score:', accuracy_score(y_test,dec_y_pred))
```

```python
# Plotting Confusion Matrix
```

```python
cm = metrics.confusion_matrix(y_test, dec_y_pred)
print(cm)
```

```python
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cm), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
print(classification_report(y_test, dec_y_pred))
```

```python
#Importing Required Libraries
import pandas as pd
import sklearn as sklearn
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import confusion_matrix,classification_report
```

```python
#Extracting the data from .csv file
data = pd.read_csv('CE802_P2_Data.csv')
```

```python
#Replacing the missing values using mean of the existing value
data.fillna(data.mean(),inplace=True)
data
```

```python
#Obtaining Class values
y= data.iloc[:,-1:]
y
```

```python
#Obtaining feature values from F1 to F15
x=data.iloc[:,:-1]
x
```

```python
print(x.shape)
print(y.shape)
```

```python
#Splitting the data into test and train
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Creating a instance of Logistic regression classifier
lr = LogisticRegression()
```

```python
#Training the model using train data
lr.fit(X_train, y_train)
```

```python
#Predicting the output for X_test
logi_y_pred = lr.predict(X_test)
```

```python
#Finding the accuracy score
print('Accuracy Score:', accuracy_score(y_test, logi_y_pred))
```

```python
# Plotting Confusion Matrix
```

```python
cm = metrics.confusion_matrix(y_test, logi_y_pred)
print(cm)
```

```python
print(classification_report(y_test, logi_y_pred))
```

```python
#Importing Required Libraries
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import sklearn as sklearn
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import confusion_matrix,classification_report
```

```python
#Extracting the data from .csv file
data = pd.read_csv('CE802_P2_Data.csv')
data
```

```python
#Replacing the missing values using mean of the existing value
data.fillna(data.mean(),inplace=True)
data
```

```python
#Obtaining Class values
y= data.iloc[:,-1].to_numpy()
y =pd.DataFrame(y)
```

```python
#Obtaining feature values from F1 to F15
x=data.iloc[:,:-1].to_numpy()
x
```

```python
print(x.shape)
print(y.shape)
```

```python
#Splitting the data into test and train
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.22, random_state=42)
```

```python
#Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Creating a instance of Random forest classifier
clf=RandomForestClassifier()
#Training the model using train data
clf.fit(X_train, y_train)
```

```python
#Predicting the output for X_test
rand_y_pred = clf.predict(X_test)
```

```python
clf.score(X_train, y_train)
```

```python
#Reshaing the y_test 
rand_y_pred = rand_y_pred.reshape(-1,1)
```

```python
y_test.shape
```

```python
#Finding the accuracy score
print('Accuracy Score:', accuracy_score(y_test,rand_y_pred))
```

```python
cm = metrics.confusion_matrix(y_test, rand_y_pred)
print(cm)
```

```python
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cm), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
print(classification_report(y_test, rand_y_pred))
```
